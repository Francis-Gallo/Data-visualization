```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df =pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df.info()
```

```python
df['species'] = iris.target

# Map species numbers to species names
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Create a scatter plot (sepal length vs sepal width)
plt.figure(figsize=(8, 6))
for species in df['species'].unique():
    species_data = df[df['species'] == species]
    plt.scatter(species_data['sepal length (cm)'], species_data['sepal width (cm)'], label=species)

# Add labels and title
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width - Iris Dataset')
plt.legend()

# Show the plot
plt.show()
```

```python
f, ax = plt.subplots(1,1, figsize=(10, 5))

sns.scatterplot(x="sepal length (cm)", y="sepal width (cm)", hue='species', data=df, ax=ax);
```

```python
df = df.rename(columns={"sepal length (cm)": "sepal_length",
                        "sepal width (cm)": "sepal_width",
                        "petal width (cm)": "petal_width"})
fig = px.scatter_3d(df, x="sepal_length", y="sepal_width", z="petal_width", color="species", size="sepal_width")
fig.show()
```

```python
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
plt.figure(figsize=(8, 6))
plt.hist(df_iris['sepal length (cm)'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length - Iris Dataset')
plt.show()
```

```python
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
plt.figure(figsize=(8, 6))
sns.histplot(df_iris['sepal length (cm)'], bins=20, color='skyblue', kde=False)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length - Iris Dataset')
plt.show()
```

```python
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
fig = px.histogram(df_iris, x='sepal length (cm)', nbins=20, title='Histogram of Sepal Length - Iris Dataset', color_discrete_sequence=['skyblue'])
fig.update_layout(
    xaxis_title='Sepal Length (cm)',
    yaxis_title='Frequency',
)
fig.show()
```

```python
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
plt.figure(figsize=(8, 6))
plt.boxplot(df_iris['sepal length (cm)'], patch_artist=True, boxprops=dict(facecolor='skyblue'))
plt.ylabel('Sepal Length (cm)')
plt.title('Box Plot of Sepal Length - Iris Dataset')
plt.show()
```

```python
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='sepal length (cm)', data=df_iris, palette='coolwarm')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.title('Box Plot of Sepal Length by Species - Iris Dataset')
plt.show()
```

```python
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
fig = px.box(df_iris, x='species', y='sepal length (cm)', title='Box Plot of Sepal Length by Species - Iris Dataset', color='species')
fig.show()
```

```python
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
species_counts = df_iris['species'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'coral'])
plt.title('Species Distribution in the Iris Dataset')
plt.show()
```

```python
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
species_counts = df_iris['species'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel", 3))
plt.title('Species Distribution in the Iris Dataset')
plt.show()
```

```python
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
species_counts = df_iris['species'].value_counts().reset_index()
species_counts.columns = ['species', 'count']
fig = px.pie(species_counts, values='count', names='species', title='Species Distribution in the Iris Dataset')
fig.show()
```

```python
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
species_counts = df_iris['species'].value_counts()
plt.figure(figsize=(8, 6))
plt.bar(species_counts.index, species_counts.values, color=['lightblue', 'lightgreen', 'coral'])
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Species Distribution in the Iris Dataset')
plt.show()
```

```python
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
species_counts = df_iris['species'].value_counts().reset_index()
species_counts.columns = ['species', 'count']
fig = px.bar(species_counts, x='species', y='count', title='Species Distribution in the Iris Dataset', color='species',
             color_discrete_sequence=px.colors.qualitative.Plotly)
fig.show()
```
